### Test de Chi

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2

def test_chi2(serie1, serie2):
    
    alpha = 0.03    
    
    
    tab_contingence = pd.crosstab(serie1.array, serie2.array)
    stat_chi2, p, dof, expected_table = chi2_contingency(tab_contingence.values)
    print('chi2 : {0:.5f},\np : {0:.5f},\ndof : {0:.5f}\n'.format(stat_chi2, p, dof))
    
    critical = chi2.ppf(1-alpha, dof)
    if p <= alpha:
        print('Variables non ind√©pendantes (H0 rejet√©e) car p = {} <= alpha = {}'.format(p, alpha))
        return False
    
    else:
        print('H0 non rejet√©e car p = {} >= alpha = {}'.format(p, alpha))
        return True
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[numeric_columns].dropna())

pca = PCA(n_components=3)
pca_result = pca.fit_transform(scaled_data)

data_pca = pd.DataFrame(pca_result, columns=['PC1', 'PC2', 'PC3'])
print(data_pca.head())

data['PC1'] = data_pca['PC1']
data['PC2'] = data_pca['PC2']
data['PC3'] = data_pca['PC3']

print("Variance expliqu√©e par chaque composante : ", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data, 
    x="PC1", 
    y="PC2", 
    hue="ingredients_that_may_be_from_palm_oil_n",
    palette="viridis", # Changer la palette ex rouge et jaune
    alpha=0.7
)

plt.xlabel("Composante Principale 1 (PC1)")
plt.ylabel("Composante Principale 2 (PC2)")
plt.title("Projection PCA avec classification par pr√©sence d'huile de palme")
plt.legend(title="Pr√©sence d'huile de palme")
plt.show()
```

```python
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

feature_coords = pd.DataFrame(loadings, index=numeric_columns, columns=['PC1', 'PC2', 'PC3'])

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='dotted')
ax.add_patch(circle)

for i, feature in enumerate(numeric_columns):
    plt.arrow(0, 0, feature_coords["PC1"][i], feature_coords["PC2"][i],
              head_width=0.05, head_length=0.05, color='red')
    plt.text(feature_coords["PC1"][i], feature_coords["PC2"][i], feature, fontsize=12, color='blue')

plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Cercle des Corr√©lations (PC1 vs PC2)")
plt.grid()
plt.show()
```
